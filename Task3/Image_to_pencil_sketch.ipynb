# **Image to Pencil Sketch with python**
*Here i am using OpenCV library for image to pencil sketch project.It can be used by pip command.pip install opencv-python*

## **Importing OpenCV library**

```python
import cv2
from google.colab.patches import cv2_imshow
```

## **Import Image from local device into COLAB**

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
image = cv2.imread("dog.jpg",cv2.IMREAD_UNCHANGED)
```

## **Showing Orginal Image**

```python
cv2_imshow(image)
cv2.waitKey(0)
```

## **Convert the orginal(RGB) image into greyscale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

## **Invert the grayscale image**

```python
inverted_image = 255-gray_image
cv2_imshow(inverted_image)
cv2.waitKey(0)
```

## **Blur the image using Gaussian Function in OpenCV**

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21,21), 0)
cv2_imshow(blurred_image)
cv2.waitKey(0)
```

## **Final step, Invert the Blurred image**

```python
final_image = 255-blurred_image
pencil_sketch = cv2.divide(gray_image, final_image,scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
